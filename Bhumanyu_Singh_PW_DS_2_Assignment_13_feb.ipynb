#### Q1. Explain why we have to use the Exception class while creating a Custom Exception.

In Python, when we create a custom exception, we typically inherit from the built-in Exception class. The Exception class serves as the base class for all the exceptions in Python, and it provides a set of methods and attributes that are essential for proper exception handling.

Here are the reasons why we use the Exception class as the base class for custom exceptions:

Consistency: By inheriting from the Exception class, our custom exception becomes consistent with other built-in exceptions in Python. This allows us to use our custom exception in a similar way as we would use any other exception in Python.

Proper Exception Handling: The Exception class provides methods like __str__() and __repr__() that help in formatting the exception message and displaying useful information when the exception is raised. By inheriting from the Exception class, our custom exception can leverage these methods, making it easier for developers to handle and understand the exception.

Customization: By creating a custom exception class, we can define additional attributes and methods that are specific to our application or use case. We can add extra context or information to the exception, making it more informative and helpful for debugging.

Catching Specific Exceptions: When we raise a custom exception, we can catch it specifically using except blocks, which allows us to handle different exceptions differently based on their types. This level of granularity in exception handling is essential for writing robust and error-tolerant code.

ImportError : This error happens when you are trying to access the module that are not found in python

ImportError is an exception that is raised when there is an error in importing a module or a name in a module. It usually occurs when the Python interpreter cannot find the specified module or when there is a problem with the module's code.

ModuleNotFoundError is a specific subclass of ImportError that is raised when the Python interpreter cannot find the specified module. It was introduced in Python 3.6 to provide a more specific error message when a module is not found. Before Python 3.6, an ImportError was raised in such cases.

#### Q6. List down some best practices for exception handling in python.

Below are all the main points:-

1. Use Specific Exceptions: Instead of catching a generic `Exception`, try to catch specific exceptions that you know might be raised. This helps in providing more meaningful error messages and allows you to handle different exceptions differently.

2. Avoid Bare Excepts: Avoid using bare `except` statements as they catch all exceptions, including system exit signals. It's better to catch specific exceptions or use `except Exception as e` to capture the exception and handle it appropriately.

3. Use `finally`: When necessary, use the `finally` block to ensure that certain code is executed whether an exception is raised or not. For example, closing a file or releasing resources.

4. Don't Ignore Exceptions: Never ignore exceptions. At the very least, log the exception so that you can track and diagnose any issues that might occur.

5. Handle Exceptions Locally: Only catch exceptions that you can handle locally. If you can't handle an exception properly, it's better to let it propagate up the call stack.

6. Use Context Managers: For resources like files, database connections, or network connections, use context managers (`with` statement) to ensure proper cleanup and resource release.

7. Avoid Deep Nested `try-except` Blocks: Avoid deeply nesting `try-except` blocks as it can make the code harder to read and maintain. Refactor the code into smaller functions and handle exceptions at appropriate levels.

8. Raising Custom Exceptions: Create custom exception classes when you need to raise specific exceptions for your application. This can help in better organization and error reporting.

9. Use Exception Chaining: When re-raising an exception, use the `from` keyword to chain exceptions together. This way, you can see the original cause of the exception.

10. Keep Error Messages Clear and Informative: When handling exceptions, provide clear and informative error messages. This will help in identifying the issue and debugging the code.

Remember, exception handling is an essential part of writing robust and reliable code. Properly handling exceptions can prevent crashes, improve error reporting, and make debugging easier.
